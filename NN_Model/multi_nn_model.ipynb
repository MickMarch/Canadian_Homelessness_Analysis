```python
import pandas as pd
from pathlib import Path
```

```python
crime_df = pd.read_csv(Path("../cleaned_data_2015_2018/all_data_merged_cleaned_2015_2018.csv"))
crime_df
```

```python
daily_crime = crime_df.groupby(by=['date', "occ_dow"])["crime"].value_counts().unstack(fill_value=0)
daily_crime
```

```python
daily_crime.columns
```

```python
target_columns = [f'{column}_count' for column in daily_crime.columns]

daily_crime.columns = target_columns
daily_crime
```

```python
daily_crime.reset_index()
```

```python
daily_weather_df = pd.read_csv(Path('../cleaned_data_2015_2018/toronto_daily_weather_2015_2018.csv'))
daily_weather_df
```

```python
crime_weather_df = daily_weather_df.merge(daily_crime, how="left", on="date")
crime_weather_df
```

```python
# Split the data into input and output variables
output_count = len(target_columns)

X = crime_weather_df.iloc[:, :-output_count].drop("date", axis=1).values
y = [crime_weather_df[target].values for target in target_columns]

assert X.shape[0] == y[0].shape[0], "Number of samples in X and y do not match"
```

```python
print(X.shape)
for target in y:
    print(target.shape)
```

```python
import numpy as np
y_stacked = np.column_stack(y)
y_stacked
```

```python
# WINNER method to unstack

y_unstack = np.array_split(y_stacked, y_stacked.shape[0])
y_unstack
```

```python
# Split data into train and test data
from sklearn.model_selection import train_test_split

# Stack the outputs (y) to make the shape match the features (X)
import numpy as np
y_stacked = np.column_stack(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_stacked, random_state=0)

assert X_train.shape[0] == y_train.shape[0], "Number of samples in X_train and y_train do not match"
assert X_test.shape[0] == y_test.shape[0], "Number of samples in X_test and y_test do not match"
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.fit_transform(X_test)
```

```python
y_train.shape
```

```python
# Unstack the y data again

y_train_list = np.array_split(y_train, y_train.shape[0])
y_test_list = np.array_split(y_test, y_test.shape[0])
```

```python
import tensorflow as tf

# Define the input shape. Equal to the amount of feature columns (Total column count - output count - removed date column)
input_shape = (len(crime_weather_df.columns) - output_count - 1, )

# Define the input layer
inputs = tf.keras.Input(shape=input_shape)

# Define the hidden layers
hidden1 = tf.keras.layers.Dense(32, activation='relu')(inputs)
hidden2 = tf.keras.layers.Dense(16, activation='relu')(hidden1)

# Define the output layers
output_layers=[]

for target_col in target_columns:
    output_layers.append(tf.keras.layers.Dense(1, activation="linear", name=target_col)(hidden2))

# Define the model
model = tf.keras.Model(inputs=inputs, outputs=output_layers)

# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
# Fit the model
model.fit(X_train_scaled, y_train, epochs=50, batch_size=32)
```

```python
# Evaluate the model on test data
loss, accuracy = model.evaluate(X_test_scaled, y_test)

# Print the results
print(loss)
print('Test accuracy:', *accuracy)
```
